A travez de los datos podemos obtener información que no se ve de forma natural, a travez de la recopilación y el analisis de la información. La estadística sirve para tomar decisiones importantes, la estadística se divide en 2: la estadística descriptiva y la estadística inferencial.

La estadística descriptiva suele usar medidas centrales, disperciones y el apoyo gráfico con el propósito de describir la información. Su objetivo es analizar y presentar la información.
Pasos:
1- Recolección y organización de los datos: La información se convierte en datos una vez que estén estructurados, la información se organiza en tablas o gráficas para transformarlos a datos. 2- El calculo de medidas descriptivas: Se utilizan diferentes medidas, como las medidas centrales para sacar esos datos. 3- Presentación de los datos: Se presentan de forma visual atractiva y fácil de entender (gráficas, diagrama). 4- Identificación de patrones y relaciones, detección de tendencias, usa la tercia central: Media, mediana, moda, media geometricá. Usa medidas de dispersión como el rango, la varianza, la desviación estandar y el coeficiente de variación y tambien usa la medidas de posicionamiento: deciles, cuartiles y los percentiles.

La estadística inferencial trabaja a travez de modelos de predicción, estudia los eventos, puedes ver como se maneja algo atravez del comportamiento, usa las probabilidades, la hipotesis y el muestreo. Se centra en realizar predicciones y tomar decisiones.
Pasos:
1- Estimación de parametros, estimación proporcionada por la muestra.
2- Conjetura de hipotesis y su prueba, nos permiten tomar decisiones sobre la validez de la población.
3- Intervalos de confianza: El futuro es impredecible, pero podemos generar una medida que nos permita pensar si tiene más probabilidad de funcionar la hipotesis.
4- Modelado estadistico: Estos modelos permiten realizar predicciones sobre el futuro

* La probabilidad: estudia los eventos.
* Las hipotesis: es una idea que no está confirmadas y son muy pocas las que se convierten a teoría, las ciencias sociales son las que más usan la estadística ya que muestran todo lo que representan la sociedad.
